# Artificial Neural Network - Regression 

### Importing the libraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
tf.__version__
```

## Part 1 - Data Preprocessing

### Importing the dataset

```python
dataset = pd.read_csv('insurance.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct1 = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct1.fit_transform(X))
X[0]
```

```python
ct2 = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [6])], remainder='passthrough')
X = np.array(ct2.fit_transform(X))
X[0]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(X[:,9])
X[:,9] = le.fit_transform(X[:,9])
```

```python
X = np.asarray(X).astype(np.float32)
X[0]
```

### Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
X_train
```

```python
(X[0,-1])
```

## Part 2 - Building the ANN

### Initializing the ANN

```python
ann = tf.keras.models.Sequential()
```

### Adding the input layer and the first hidden layer

```python
ann.add(tf.keras.layers.Dense(units=120, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=60, activation='relu'))
ann.add(tf.keras.layers.Dense(units=60, activation='relu'))
```

### Adding the second hidden layer

```python
ann.add(tf.keras.layers.Dense(units=60, activation='relu'))
```

### Adding the output layer

```python
ann.add(tf.keras.layers.Dense(units=1))
```

## Part 3 - Training the ANN

### Compiling the ANN

```python
ann.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

### Training the ANN model on the Training set

```python
ann.fit(X_train, y_train, batch_size = 32, epochs = 1000)
```

### Predicting the results of the Test set

```python
import sklearn.metrics as sm
y_pred = ann.predict(X_test)
np.set_printoptions(precision=2)
print(round(sm.r2_score(y_test, y_pred), 2))
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```
